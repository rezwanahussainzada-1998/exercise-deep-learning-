### TASK 2: Autocovariance
Plot the autocovariance for a and b

```python
# Calculate the autocovariance for a and b
autocov_a = np.correlate(a, a, mode='full')
autocov_b = np.correlate(b, b, mode='full')

# Plot the autocovariance for a
plt.figure()
plt.plot(autocov_a)
plt.title('Autocovariance for a')
plt.xlabel('Lag')
plt.ylabel('Autocovariance')
plt.show()

# Plot the autocovariance for b
plt.figure()
plt.plot(autocov_b)
plt.title('Autocovariance for b')
plt.xlabel('Lag')
plt.ylabel('Autocovariance')
plt.show()
```

```python
a = EEG_data[0][0:2500]
dt = (1500/500) - 0  # Define the sampling interval
N = a.shape[0]    # Define the total number of data points
T = N * dt        # Define the total duration of the data
lags = np.arange(-len(a) + 1, len(a))    # Compute the lags for the full autocovariance vector
                                      # ... and the autocov for L +/- 100 indices
ac = 1 / N * np.correlate(a - a.mean(), a - a.mean(), mode='full')
inds = abs(lags) <= 100               # Find the lags that are within 100 time steps
plt.figure(figsize=(18 , 8))
plt.plot(lags[inds] * dt, ac[inds])       # ... and plot them
plt.xlabel('Lag [ms]')                     # ... with axes labelled
plt.ylabel('Autocovariance')
plt.title('Autocovariance for a')
plt.show()
```

```python
b = EEG_data[27][3000:4500]
dt = (4500/500) - (3000/500)  # Define the sampling interval
N = b.shape[0]    # Define the total number of data points
T = N * dt        # Define the total duration of the data
lags = np.arange(-len(b) + 1, len(b))    # Compute the lags for the full autocovariance vector
                                      # ... and the autocov for L +/- 100 indices
ac = 1 / N * np.correlate(b - b.mean(), b - b.mean(), mode='full')
inds = abs(lags) <= 100               # Find the lags that are within 100 time steps
plt.figure(figsize=(18 , 8))
plt.plot(lags[inds] * dt, ac[inds])  # ... and plot them
plt.xlabel('Lag [ms]')                     # ... with axes labelled
plt.ylabel('Autocovariance')
plt.title('Autocovariance for b')
plt.show()
```

### TASK 3 : Spectrogram


```python
from scipy.signal import spectrogram
```

```python
# plot spectrograms for nfft = 500, 1s

a = EEG_data[0][0:2500]
nfft = 500
overlap = 2
fs = 500
#plt.figure(figsize=(10 , 10))
freqs , times, Sxx = spectrogram(a, fs=fs, nperseg=nfft, noverlap= overlap)
plt.pcolormesh(times, freqs, 10*np.log10(Sxx))
plt.colorbar()
plt.xlabel('Time [s]')
plt.ylabel('Frequency [Hz]')
```
